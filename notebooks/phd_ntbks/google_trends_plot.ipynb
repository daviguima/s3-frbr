```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('google.xlsx', sheet_name=0)
```

```python
df.head()
```

```python
plt.rcParams['figure.figsize'] = [12, 4]
plt.plot(df['Month'],df['Remote sensing'],label='Remote sensing')
plt.plot(df['Month'],df['Big data'],label='Big data')
plt.plot(df['Month'],df['Machine learning'],label='Machine learning')
plt.plot(df['Month'],df['Deep learning'],label='Deep learning')
plt.plot(df['Month'],df['GPU'],label='GPU')
plt.title('Google Trends: keyword popularity over time.')
plt.ylabel('Popularity (%)')
plt.legend()
plt.savefig('google_trends.png', dpi=100)
```
