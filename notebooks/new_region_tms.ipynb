```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import sys
sys.path.append('../')

from tsgen import TsGenerator
```

```python
tsg = TsGenerator()
```

```python
workdir = "D:\\processing\\win"

ts_list = ["coari-ts.csv",
          "fonte-boa-ts.csv",
          "madeira-ts.csv",
          "manacapuru-ts.csv",
          "negro-ts.csv",
          "tefe-ts.csv"]

coari_dict = pd.read_csv(os.path.join(workdir,ts_list[0]), parse_dates=['Datetime'])
fonte_boa_dict = pd.read_csv(os.path.join(workdir,ts_list[1]), parse_dates=['Datetime'])
madeira_dict = pd.read_csv(os.path.join(workdir,ts_list[2]), parse_dates=['Datetime'])
manacapuru_dict = pd.read_csv(os.path.join(workdir,ts_list[3]), parse_dates=['Datetime'])
negro_dict = pd.read_csv(os.path.join(workdir,ts_list[4]), parse_dates=['Datetime'])
tefe_dict = pd.read_csv(os.path.join(workdir,ts_list[5]), parse_dates=['Datetime'])
```

```python
coari_dict.keys()
```

```python
tsg.bname_dict
```

```python
plt.rcParams['figure.figsize'] = [16, 6]
# fig = plt.figure()
ax = plt.axes()
ax.set_title('S3-WFR Reflectance time-series from 2019-03-09 to 2020-03-31 over Oa08:665 nm', fontsize=16)

ax.plot(coari_dict['Datetime'], coari_dict['B8-665'], marker='o', markersize=2, linewidth=0.5, label='Coari')
ax.plot(fonte_boa_dict['Datetime'], fonte_boa_dict['B8-665'], marker='o', markersize=2, linewidth=0.5, label='Fonte Boa')
ax.plot(madeira_dict['Datetime'], madeira_dict['B8-665'], marker='o', markersize=2, linewidth=0.5, label='Madeira')
ax.plot(manacapuru_dict['Datetime'], manacapuru_dict['B8-665'], marker='o', markersize=2, linewidth=0.5, label='Manacapuru')
ax.plot(negro_dict['Datetime'], negro_dict['B8-665'], marker='o', markersize=2, linewidth=0.5, label='Negro')
ax.plot(tefe_dict['Datetime'], tefe_dict['B8-665'], marker='o', markersize=2, linewidth=0.5, label='Tefe')

ax.set_xlabel('Date', fontsize=16)
ax.set_ylabel('Reflectance', fontsize=16)
ax.legend()
plt.savefig('all-regions-ts-plot-b8.png', dpi=300)
plt.show()
```
