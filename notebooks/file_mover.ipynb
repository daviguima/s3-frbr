```python
import os
import shutil
```

```python
dest = 'D:\\S3\\L2_WFR'

source = 'D:\\S3\\L2_WFR\\xtrct'

files = os.listdir(source)

len(files)
```

```python
for f in files:
    print(f'moving file: {f} to {dest}')
    current_file = os.path.join(os.path.join(source,f),f)
    shutil.move(current_file,dest)
    print('done.')
```
